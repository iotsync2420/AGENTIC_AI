```python
#import libraries
import google.generativeai as genai
import ipywidgets as widgets
from IPython.display import display, Markdown
```

```python
#setup gemini api
API_KEY= "AIzaSyBlG2qRk0z52kNKUkzjh6IMpahFGl2C4J8"
genai.configure(api_key=API_KEY)
model=genai.GenerativeModel("gemini-2.5-flash")
```

```python
#input form

input=widgets.Text(
    description="enter your prompt ",
    placeholder="eg: AI/ML",
    layout=widgets.Layout(width="400px")
)
tone_inputs= widgets.Dropdown(
    options=['motivation', 'academic ','emotional'],
    value='motivation',
    description='tone'
)
hastag_input=widgets.Text(
    description="enter hashtags",
    placeholder="eg: #AI #ML",
    layout=widgets.Layout(width="400px")
)
audience_input=widgets.Text(
    description="audience",
    placeholder="eg: college students",
    layout=widgets.Layout(width="400px")
)
submit_button=widgets.Button(description="generate tweet",buttonstyle='primary',tooltips='click to generate tweet')
Output=widgets.Output()
```

```python
#geerate tweet function
def generate_tweet(b):
  Output.clear_output()
  prompt= f"""
  YOU ARE A GR8 CONTENT WRITER !
  Generate a tweet on topic "{input.value}",
  use a "{tone_inputs.value.lower()}"tone,
  Generate a tweet for audience "{audience_input.value}",
  use hashtags "{hastag_input.value}"


  ALL THE VERY BEST !
  (KEEP IT UNDER 200 WORDS )
  """

  with Output:
    try:
      response=model.generate_content(prompt)
      tweet= response.text.strip()
      display(Markdown(f'''generated tweet :{tweet}'''))
    except Exception as e :
      print("error ",e )
submit_button.on_click(generate_tweet)
```

```python
#form display
form = widgets.VBox([

         widgets.HTML(value="<h3> Tweet Generator Agent</h3>"),

        Output,input,tone_inputs,hastag_input,audience_input,submit_button])

display(form)
```
